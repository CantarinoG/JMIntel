# 01 - Data Inspection

In this notebook, we connect to the `jobs.db`database, load the scraped job listings, and perform an initial sanity check on the data quality.

```python
import sqlite3
import pandas as pd
import os
```

## Connecting to the Database
We are loading the data from the SQLite database into a Pandas DataFrame for analysis.

```python
db_path = os.path.join("..", "jobs.db")
conn = sqlite3.connect(db_path)

df = pd.read_sql_query("SELECT * FROM jobs", conn)
conn.close()
```

## Data Preview
Let's look at the first few rows to ensure the columns were populated correctly by the scraper and the AI.

### Fields Description
| Column Name | Description | Data Type |
| :--- | :--- | :--- |
| **id** | Unique identifier for each database record. | Integer |
| **title** | The official job title as posted by the employer. | Text |
| **description** | The full text of the job description. | Text |
| **company** | The name of the hiring organization. | Text |
| **url** | The web address of the original job posting. | Text |
| **platform** | The source platform (e.g., LinkedIn). | Text |
| **date_collected** | The date and time when the job was scraped. | Date/Time |
| **job_category** | The specific field (Frontend, Data, DevOps, etc.). | Text |
| **seniority_level** | Experience tier (Junior, Mid, Senior, Lead). | Text |
| **employment_type** | Full-time, Part-time, Internship, etc. | Text |
| **contract_duration** | Duration if it's a temporary or project-based role. | Text |
| **remote_type** | Remote, Hybrid, or On-site. | Text |
| **language** | Primary language required for the job. | Text |
| **monthly_salary_min_brl**| Estimated minimum monthly salary in BRL. | Number |
| **monthly_salary_max_brl**| Estimated maximum monthly salary in BRL. | Number |
| **experience_years_min** | Minimum years of experience requested. | Number |
| **hard_skills** | List of technical skills/tools (stored as JSON). | List/JSON |
