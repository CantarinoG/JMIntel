# 03 - Descriptive Analysis
In this notebook, we transform our cleaned dataset into actionable insights. We will perform descriptive analysis to understand the current tech market trends.

```python
import pandas as pd
import plotly.express as px
import os
```

```python
processed_data_path = os.path.join("..", "data", "processed", "clean_jobs.pkl")
df = pd.read_pickle(processed_data_path)
```

## Hard Skill Demand
The first step in our "Intelligence" extraction is identifying which tecnhologies are most frequent accross all job postings. This tells us what the market considers "Must-Have" skills.

```python
skills_number = 15
all_skills = df['hard_skills'].explode()

top_skills = all_skills.value_counts().head(skills_number).reset_index()
top_skills.columns = ['Skill', 'Count']

fig = px.bar(
    top_skills, 
    x='Count', 
    y='Skill', 
    orientation='h',
    title='Top ' + str(skills_number) + ' Most Demanded Tech Skills',
    labels={'Count': 'Number of Job Postings'},
    color='Count',
    color_continuous_scale='Viridis'
)

fig.update_layout(yaxis={'categoryorder':'total ascending'})
fig.show()
```

## Role Distribution
The goal is to identifying which roles have higher demand.

```python
category_counts = df['job_category'].value_counts().head(10).reset_index()
category_counts.columns = ['Job Category', 'Count']

fig_category = px.pie(
    category_counts, 
    values='Count',
    names='Job Category', 
    title='Market Composition: Top 10 Most Frequent Tech Roles',
    hole=0.4,
    color_discrete_sequence=px.colors.qualitative.Set3
)

fig_category.update_traces(textinfo='percent+label')
fig_category.show()
```

## Experience Requirements
The goal is to identify the distribution between different seniority positions.

```python
seniority_order = ['intern', 'junior', 'midlevel', 'senior', 'lead', 'not specified']
fig_seniority = px.bar(
    df['seniority_level'].value_counts().reindex(seniority_order).reset_index(),
    x='seniority_level',
    y='count',
    title='Market Seniority Requirements',
    labels={'seniority_level': 'Level', 'count': 'Number of Jobs'},
    color='count',
    color_continuous_scale='Blues'
)
fig_seniority.show()
```

## Company Activity
The goal is to identify top hiring companies.

```python
top_companies = df['company'].value_counts().head(10).reset_index()
top_companies.columns = ['Company', 'Job Count']
fig_companies = px.bar(
    top_companies,
    x='Job Count',
    y='Company',
    orientation='h',
    title='Top 10 Most Active Hiring Companies',
    color='Job Count',
    color_continuous_scale='Reds'
)
fig_companies.update_layout(yaxis={'categoryorder':'total ascending'})
fig_companies.show()
```

## Junior and Mid-level Market Snapshot
To better understand the requirements for entry and intermediate roles, we will filter our dataset to include only 'Junior' and 'Midlevel' seniority.

This helps us identify which technologies and roles are most acessible at these career stages.

```python
target_levels = ['junior', 'midlevel']
df_jr_mid = df[df['seniority_level'].isin(target_levels)]

fig_roles_jr = px.pie(
    df_jr_mid, 
    names='job_category', 
    title='Role Distribution: Junior & Mid-Level Only',
    hole=0.4,
    color_discrete_sequence=px.colors.qualitative.Pastel
)
fig_roles_jr.update_traces(textinfo='percent+label')
fig_roles_jr.show()
```

```python
jr_skills = df_jr_mid['hard_skills'].explode()
top_jr_skills = jr_skills.value_counts().head(15).reset_index()
top_jr_skills.columns = ['Skill', 'Count']
fig_jr_skills = px.bar(
    top_jr_skills, 
    x='Count', 
    y='Skill', 
    orientation='h',
    title='Top 15 Skills for Junior & Mid-Level Roles',
    labels={'Count': 'Number of Job Postings'},
    color='Count',
    color_continuous_scale='Viridis'
)
fig_jr_skills.update_layout(yaxis={'categoryorder':'total ascending'})
fig_jr_skills.show()
```

```python
top_jr_companies = df_jr_mid['company'].value_counts().head(10).reset_index()
top_jr_companies.columns = ['Company', 'Job Count']

fig_jr_companies = px.bar(
    top_jr_companies,
    x='Job Count',
    y='Company',
    orientation='h',
    title='Top 10 Companies Hiring Junior & Mid-Level Roles',
    color='Job Count',
    color_continuous_scale='Magma'
)
fig_jr_companies.update_layout(yaxis={'categoryorder':'total ascending'})
fig_jr_companies.show()
```
